```python
from phySyncFirmata import ArduinoNano
import phySyncFirmata
import time
```

```python
PORT = ArduinoNano.AUTODETECT
board = ArduinoNano(PORT)
# setting port for Arduino

board.samplingOn()
```

```python
t = 0.1
frequency = 0.1
count = 1
```

```python
SOS = '''SOS
'''
```

```python
Buzzer = board.get_pin('d:5:p')
```

```python
while True:
    print(str(count) + 'st time')
    for element in SOS:
        
        if element == 'O':
            print('O: ---')
            Buzzer.write(frequency)
            board.pass_time(3*t)
            Buzzer.write(0)
            board.pass_time(t)
            Buzzer.write(frequency)
            board.pass_time(3*t)
            Buzzer.write(0)
            board.pass_time(t)
            Buzzer.write(frequency)
            board.pass_time(3*t)
            Buzzer.write(0)
            board.pass_time(3*t)
        
        elif element == 'S':
            print('S: ...')
            Buzzer.write(frequency)
            board.pass_time(t)
            Buzzer.write(0)
            board.pass_time(t)
            Buzzer.write(frequency)
            board.pass_time(t)
            Buzzer.write(0)
            board.pass_time(t)
            Buzzer.write(frequency)
            board.pass_time(t)
            Buzzer.write(0)
            board.pass_time(3*t)
    count += 1
```
